```python
import keyboard
import mss
import cv2
import numpy
from time import time, sleep
import pyautogui
from PIL import Image as im


wood_left=cv2.imread('wood_left.png')

wood_right=cv2.imread('wood_right.png')



w = wood_left.shape[1]
h = wood_left.shape[0]




from time import time, sleep




import time as time



dimensions_left = {
        'left': 730,
        'top': 460,
        'width': 200,
        'height': 200
    }

dimensions_right = {
        'left': 1030,
        'top': 460,
        'width': 200,
        'height': 200
    }

sct = mss.mss()
time.sleep(5)
left = True
while True:
    

    
    if left:
        scr = numpy.array(sct.grab(dimensions_left))
        wood = wood_left
    else:
        scr = numpy.array(sct.grab(dimensions_right))
        wood = wood_right

        
    scr_remove = scr[:,:,:3]
    
    result = cv2.matchTemplate(scr_remove, wood, cv2.TM_CCOEFF_NORMED)
    
    
    _, max_val, _, max_loc = cv2.minMaxLoc(result)
    print(f"Max Val: {max_val} Max Loc: {max_loc}")
    src = scr.copy()
    x=0
    if left is False:
        if max_val < .70:
            left = False
        else:
            left=True
        
    elif left is True:     
        if max_val < .70:
            left = True
        else:
            left=False
    if left:
        x=340
    else:
        x=1200
            
            
    pyautogui.click(x=x, y=518)
    sleep(0.1)
    if keyboard.is_pressed('k'):
        break      
```
